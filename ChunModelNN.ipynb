```python
import pandas as pd

df = pd.read_csv("datasets/Churn_Modelling.csv", header=0)
df.head()
```

```python
df = df.drop(["RowNumber", "CustomerId", "Surname"], axis=1)

ageMap = { "Female": 0, "Male": 1}

df["Gender"] = df["Gender"].map(ageMap)
df.head()
```

```python
dummies_df = pd.get_dummies(df["Geography"], drop_first=True)
df = pd.concat([df, dummies_df], axis=1)
df.drop(["Geography"], axis=1, inplace=True)
df.head()
```

```python
X = df.iloc[:,:9]
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(4,3), max_iter=400)
mlp.fit(X_train, y_train)
```

```python
mlp.score(X_test, y_test)
```

```python
scores = []
neurons = []
for i in range(1, 10):
    mlp = MLPClassifier(hidden_layer_sizes=(i))
    mlp.fit(X_train, y_train)
    score = mlp.score(X_test, y_test)
    scores.append(score)
    neurons.append(i)

import matplotlib.pyplot as plt
plt.plot(neurons, scores)
plt.show()
```
