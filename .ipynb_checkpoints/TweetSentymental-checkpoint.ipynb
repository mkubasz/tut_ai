```python
import pandas as pd
df = pd.read_csv('datasets/Tweets.csv', header=0, sep=',')
df.head()
```

```python
df = df[['airline_sentiment', 'text']]
df.head()
```

```python
import re

def norm1(row):
    return re.sub("[-()\"#/@;:<>{}`+=-|.!?,]", '', row['norm'])

df['norm'] = df.apply(norm1, axis=1)
df.head()
```
