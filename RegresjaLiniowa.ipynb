```python
# wczytanie danych
import pandas as pd
df = pd.read_csv("diamonds_single.csv", sep=";", header=0)
df.head()
df.shape
```

```python
# wizualizację danych
%matplotlib inline
df.plot.scatter(x='carat', y='price')
```

```python
# podział danych
from sklearn.model_selection import train_test_split
X_diam = df[['carat']]
y_diam = df[['price']]
Xtrain, Xtest, ytrain, ytest = train_test_split(X_diam, y_diam, random_state=1)
```

```python
# przygotowanie modelu
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
# wytrenowanie modelu
regr.fit(Xtrain, ytrain)
```

```python
# predykcji na danych testowych
y_pred = regr.predict(Xtest)
```

```python
# wyniki
from sklearn.metrics import mean_squared_error
print("Współczynnik:", regr.coef_)
print("Wyraz wolny:", regr.intercept_ )
print("Błąd średniokwadratowy", mean_squared_error(ytest, y_pred))
print("R^2", regr.score(Xtrain, ytrain))
```

```python
import matplotlib.pyplot as plt
plt.scatter(Xtrain, ytrain)
plt.plot(Xtest, y_pred, color='red', linewidth=3)
```
