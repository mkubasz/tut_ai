```python
import pandas as pd
df = pd.read_csv('datasets/default_credit_card.csv', sep=";", header=0)
df.head()
```

```python
df.describe()
```

```python
X = df.drop(['default payment next month', 'SEX', 'EDUCATION', 'MARRIAGE'], axis=1)
y = df['default payment next month']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
pd.DataFrame(X_train).describe()
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(11,))
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
scores = []
neurons = []

for i in range(1, 22):
    mlp = MLPClassifier(hidden_layer_sizes=(i))
    mlp.fit(X_train, y_train)
    score = mlp.score(X_test, y_test)
    scores.append(score)
    neurons.append(i)

import matplotlib.pyplot as plt
plt.plot(neurons, scores)
plt.show()
```
